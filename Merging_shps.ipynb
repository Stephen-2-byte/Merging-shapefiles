```python
import geopandas as gpd
import pandas as pd

# Load shapefiles
gdf1 = gpd.read_file("/content/Projected_MangoBlocks.shp")
gdf2 = gpd.read_file("/content/yieldpredictedblocks.shp")

# Reproject second shapefile to first shapefile CRS
gdf2 = gdf2.to_crs(gdf1.crs)

# Merge
merged = gpd.GeoDataFrame(pd.concat([gdf1, gdf2], ignore_index=True), crs=gdf1.crs)

# Save output
merged.to_file("merged_blocks.shp")
```
